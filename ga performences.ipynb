```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
data = pd.read_csv("F:\dataP\Finalfitness.csv")

import numpy as np
data.head(7)
```

```python
from matplotlib.pyplot import figure
fige = figure(num=None, figsize=(18, 16), dpi=100, facecolor='w', edgecolor='k')
fig, ax = plt.subplots(figsize=(18,10))
ax.plot( data['best6k'], '-b', label='Best 6K')
ax.plot( data['best10k'], '-r', label='Best 10K')
ax.plot( data['best15k'], '-m', label='Best 15K')
ax.plot( data['best30k'], '-.b', label='Best 30K')
ax.plot( data['bestall'], '--r', label='Best All')

ax.legend(frameon=False, loc='lower center', ncol=2);

plt.xlabel("Generation")
plt.ylabel("Fitness Measure")
fig.savefig('test.jpg')
```
